```python
# unique categories in each categorical attribute
BCR=df['Bloomberg Composite Rating']
Categories=set(BCR)
sys.stdout.write("Unique Label Values \n")
print(Categories)
sys.stdout.write(" \n")
```

```python
sys.stdout.write("\nCounts for Each Value of Categorical Label \n")
catCount={}
for elt in BCR:
    if elt in catCount:
        catCount[elt] += 1
    else:
        catCount[elt]=1
print(catCount)
```
