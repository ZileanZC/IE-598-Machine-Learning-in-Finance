```python
X=df.iloc[:,1:23]
y=df['DEFAULT']
X.head()
```

# Random test train splits

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

accuracy_train=[]
accuracy_test=[]

SEED=np.arange(1,11)
for i in SEED:
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    # Instantiate a Decision Tree classifier: tree
    tree = DecisionTreeClassifier()
    tree.fit(X_train, y_train)
    y_pred_train=tree.predict(X_train)
    y_pred_test=tree.predict(X_test)

    accuracy_train.append(accuracy_score(y_train,y_pred_train))
    print('train_accuracy_score:',accuracy_score(y_train,y_pred_train))
    accuracy_test.append(accuracy_score(y_test,y_pred_test))
    print('test_accuracy_score',accuracy_score(y_test,y_pred_test))
    print('-------------------------------')
```

```python
train_mean=np.mean(accuracy_train)
print("train mean:", train_mean)
train_std=np.std(accuracy_train)
print("train std:", train_std)
test_mean=np.mean(accuracy_test)
print("test mean:", test_mean)
test_std=np.std(accuracy_test)
print("test std:", test_std)
```

# Cross Validation
